```python
import librosa
import librosa.display
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

from IPython.display import Audio
```

## flauta

```python
#load arquivo wav
#stereo

x_flute, sr = librosa.load('../data/instrument_sample/Flute.nonvib.ff.A4.stereo.wav', sr = None)
```

```python
Audio(data=x_flute, rate=sr)
```

```python
# formato de onda no dominio do tempo
plt.figure(figsize=(10,5))
librosa.display.waveshow(x_flute, sr =sr, color='grey')
plt.title('Flauta')
# plt.savefig('wave_big.png', bbox_inches='tight')
```

```python
x_flute.shape
```

```python
# formato de onda no dominio do tempo
plt.figure(figsize=(9,7))
librosa.display.waveshow(x_flute[50000:50300], sr =sr, color='grey')
plt.title('Flauta')
plt.savefig('../images/code_generated/inst_samples/flu_wave.png', bbox_inches='tight')
```

## sum

```python
print(len(x_vio[:105003]), len(x_flute[:105003]), len(x_sax[:105003]),  len(x_tru[:105003]))
```

```python
df = pd.DataFrame({'vio': x_vio[:105003], 'tru':x_tru[:105003], 'flu': x_flute[:105003], 'sax': x_sax[:105003]}) 
```

```python
sum_all = df.sum(axis=1)
```

```python
a = np.array(sum_all)
```

```python
Audio(data=a, rate=sr)
```

```python
# formato de onda no dominio do tempo
plt.figure(figsize=(10,5))
librosa.display.waveshow(a, sr =sr, color='grey')
plt.title('Soma')
# plt.savefig('wave_big.png', bbox_inches='tight')
```

```python
# formato de onda no dominio do tempo
plt.figure(figsize=(9,7))
librosa.display.waveshow(a[50000:50300], sr =sr, color='grey')
plt.title('Soma')
plt.savefig('../images/code_generated/inst_samples/sum_wave.png', bbox_inches='tight')
```

```python
# fig, [ax1, ax2, ax3, ax4, ax5] = plt.subplots(5, 1, sharex=True)
plt.figure(figsize=(10, 10))

plt.subplot(5, 1, 1)
librosa.display.waveshow(x_flute[50000:50400], sr =sr, color='grey')
plt.title('Flauta')

plt.subplot(5, 1, 2)
librosa.display.waveshow(x_sax[50000:50400], sr =sr, color='grey')
plt.title('Saxofone')

plt.subplot(5, 1, 3)
librosa.display.waveshow(x_tru[50000:50400], sr =sr, color='grey')
plt.title('Trompete')

plt.subplot(5, 1, 4)
librosa.display.waveshow(x_vio[50000:50400], sr =sr, color='grey')
plt.title('Violino')

plt.subplot(5, 1, 5)
librosa.display.waveshow(a[50000:50400], sr =sr, color='grey')
plt.title('Soma dos sinais')
plt.tight_layout()
plt.savefig('../images/code_generated/inst_samples/sum_subplot.png', bbox_inches='tight', dpi=600)
plt.show()
```

```python
S_flute = np.abs(librosa.stft(x_flute))
S_sax = np.abs(librosa.stft(x_sax))
S_tru = np.abs(librosa.stft(x_tru))
S_vio = np.abs(librosa.stft(x_vio))
S_a = np.abs(librosa.stft(a))
```
