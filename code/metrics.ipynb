```python
ss_values = {'model': ['SVC', 'SVC','SVC','SVC','SVC','SVC',
                       'RF', 'RF', 'RF', 'RF', 'RF', 'RF',
                       'ANN', 'ANN', 'ANN', 'ANN', 'ANN', 'ANN'],
             'label': ['Flauta', 'Guitarra', 'Piano', 'Saxofone', 'Trompete', 'Violino',
                       'Flauta', 'Guitarra', 'Piano', 'Saxofone', 'Trompete', 'Violino',
                       'Flauta', 'Guitarra', 'Piano', 'Saxofone', 'Trompete', 'Violino'],
             'recall': [65.33, 84.23, 73.2, 64.64, 69.31, 70,
                           35.33, 75.52, 74.74, 43.09, 53.44, 51.25,
                           30, 81.33, 66.49, 23.76, 33.33, 43.75]}
```

```python
precisao = pd.DataFrame.from_dict(pr_values)
sens = pd.DataFrame.from_dict(ss_values)
```

```python
fig = plt.figure(figsize = (10,5))
ax = sns.barplot(data=sens, x="label", y="recall", hue="model", palette='turbo')
plt.ylabel('Sensibilidade (%)')
ax.set(xlabel=None)
ax.legend().set_title('')
plt.savefig('../images/code_generated//models/recall.png', bbox_inches='tight', dpi = 256)
plt.show()
```
